```python
# IMPORT LIBRARIES
```

```python
import pandas as pd
import sqlite3
```

```python
# CONNECT TO THE DATABASE
```

```python
conn = sqlite3.connect('ipl_data.db')
print("DataBase connected successfully")
```

```python
# ============================================
# ALL RCB MATCHES (WHERE with OR)
# ============================================

query = """
SELECT *
FROM matches
WHERE team1 = 'Royal Challengers Bengaluru'
   OR team2 = 'Royal Challengers Bengaluru'
"""

rcb_matches = pd.read_sql_query(query, conn)
print(f"Total RCB matches: {len(rcb_matches)}")
rcb_matches.head()
```

```python
# COUNT RCB MATCHES IN SQL
```

```python
query = """
SELECT COUNT(*) AS total_matches
FROM matches
WHERE team1 = 'Royal Challengers Bengaluru'
   OR team2 = 'Royal Challengers Bengaluru'
"""

result = pd.read_sql_query(query, conn)
print(f"RCB total matches: {result['total_matches'][0]}")
```

```python
# COUNT RCB WINS
```

```python
query = """
SELECT COUNT(*) AS wins
FROM matches
WHERE winner = 'Royal Challengers Bengaluru'
"""

result = pd.read_sql_query(query, conn)
print(f"RCB wins: {result['wins'][0]}")
```

```python
# COUNT RCB LOSSES
```

```python
query = """
SELECT COUNT(*) AS losses
FROM matches
WHERE (team1 = 'Royal Challengers Bengaluru'
    OR team2 = 'Royal Challengers Bengaluru')
  AND winner != 'Royal Challengers Bengaluru'
"""

result = pd.read_sql_query(query, conn)
print(f"RCB losses: {result['losses'][0]}")
```

```python
# RCB MATCHES BY SEASON
```

```python
query = """
SELECT season, COUNT(*) AS matches_played
FROM matches
WHERE team1 = 'Royal Challengers Bengaluru'
   OR team2 = 'Royal Challengers Bengaluru'
GROUP BY season
ORDER BY season
"""

season_counts = pd.read_sql_query(query, conn)
print(season_counts)
```

```python
# TOP 5 VENUES FOR RCB MATCHES
```

```python
query = """
SELECT venue, COUNT(*) AS matches_played
FROM matches
WHERE team1 = 'Royal Challengers Bengaluru'
   OR team2 = 'Royal Challengers Bengaluru'
GROUP BY venue
ORDER BY matches_played DESC
LIMIT 5
"""

top_venues = pd.read_sql_query(query, conn)
print(top_venues)
```

```python
# CLOSE CONNECTION
```

```python
conn.close()
print("Connection closed.")
```
