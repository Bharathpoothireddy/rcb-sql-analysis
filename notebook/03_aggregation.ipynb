```python
# IMPORT LIBRARIES
```

```python
import pandas as pd
import sqlite3
```

```python
# CONNECT DATABASE
```

```python
conn = sqlite3.connect('ipl_data.db')
print("Database connected successfully")
```

```python
# TOTAL MATCHES PER SEASON
```

```python
q1 = pd.read_sql_query(" select season, count(*) as matches_played from matches where team1 = 'Royal Challengers Bengaluru' or team2 = 'Royal Challengers Bengaluru' group by season ",conn)
print("Matches per season:")
print(q1)
```

```python
# WINS PER SEASON
```

```python
q2 = pd.read_sql_query("""
    SELECT 
        season,
        COUNT(*) as matches_won
    FROM matches 
    WHERE winner LIKE '%Royal Challengers%' 
    GROUP BY season, winner
    ORDER BY season
""", conn)
print("Wins per season:")
display(q2)
```

```python
# COMBINED: MATCHES + WINS PER SEASON (CASE WHEN)
```

```python
q3 = pd.read_sql_query("""
    SELECT 
        season,
        COUNT(*) as matches_played,
        COUNT(CASE WHEN winner LIKE '%Royal Challengers%' THEN 1 END) as matches_won
    FROM matches 
    WHERE team1 = 'Royal Challengers Bengaluru' or team2 = 'Royal Challengers Bengaluru'
    GROUP BY season
    ORDER BY season
""", conn)
print("Combined season data:")
display(q3)
```

```python
# WIN PERCENTAGE PER SEASON
```

```python
q4 = pd.read_sql_query("""
    SELECT 
        season,
        COUNT(*) as matches_played,
        COUNT(CASE WHEN winner LIKE '%Royal Challengers%' THEN 1 END) as matches_won,
        ROUND(COUNT(CASE WHEN winner LIKE '%Royal Challengers%' THEN 1 END) * 100.0 / COUNT(*), 1) as win_pct
    FROM matches 
    WHERE team1 = 'Royal Challengers Bengaluru' or team2 = 'Royal Challengers Bengaluru'
    GROUP BY season
    ORDER BY season
""", conn)
print("Win percentage per season:")
display(q4)
```

```python
# BEST SEASON (HIGHEST WIN%) 
```

```python
q5 = """
SELECT 
    season,
    COUNT(*) AS total_matches,
    COUNT(CASE WHEN winner like '%Royal Challengers%' THEN 1 END) AS wins,
    ROUND(COUNT(CASE WHEN winner like '%Royal Challengers%' THEN 1 END) * 100.0 / COUNT(*), 1) AS win_pct
FROM matches
WHERE team1 = 'Royal Challengers Bengaluru'
   OR team2 = 'Royal Challengers Bengaluru'
GROUP BY season
ORDER BY win_pct DESC
LIMIT 1
"""

best_season = pd.read_sql_query(q5, conn)
print("Best season:")
print(best_season)
```

```python
# WORST SEASON (LOWEST WIN%)
```

```python
q6 = """
SELECT 
    season,
    COUNT(*) AS total_matches,
    COUNT(CASE WHEN winner like '%Royal Challengers%' THEN 1 END) AS wins,
    ROUND(COUNT(CASE WHEN winner like '%Royal Challengers%' THEN 1 END) * 100.0 / COUNT(*), 1) AS win_pct
FROM matches
WHERE team1 = 'Royal Challengers Bengaluru'
   OR team2 = 'Royal Challengers Bengaluru'
GROUP BY season
ORDER BY win_pct ASC
LIMIT 1
"""

worst_season = pd.read_sql_query(q6, conn)
print("Worst season:")
print(worst_season)
```

```python
# OVERALL WIN PERCENTAGE
```

```python
q7 = """
SELECT 
    COUNT(*) AS total_matches,
    COUNT(CASE WHEN winner like '%Royal Challengers%' THEN 1 END) AS wins,
    ROUND(COUNT(CASE WHEN winner like '%Royal Challengers%' THEN 1 END) * 100.0 / COUNT(*), 1) AS win_pct
FROM matches
WHERE team1 = 'Royal Challengers Bengaluru'
   OR team2 = 'Royal Challengers Bengaluru'
"""

overall = pd.read_sql_query(q7, conn)
print("Overall RCB record:")
print(overall)
```
