```python
# IMPORT THE LIBRARIES
```

```python
import pandas as pd
import sqlite3
```

```python
# CREATE DATABASE CONNECTIONS
```

```python
conn = sqlite3.connect('ipl_data.db')
print("Database connected successfully")
```

```python
# LOAD CSV FILES
```

```python
matches = pd.read_csv(r"H:\project\matches.csv")
deliveries = pd.read_csv(r"H:\project\deliveries.csv")

print("CSV files Extracted successfully !")
print(f"Matches: {matches.shape}")
print(f"Deliveries: {deliveries.shape}")
```

```python
# CHECK INCONSISTENCY TEAM NAMES
```

```python
print(pd.concat([matches['team1'], matches['team2']]).unique())
```

```python
# STANDARDIZE TEAM NAMES
```

```python
matches.replace({
    'Royal Challengers Bangalore' : 'Royal Challengers Bengaluru',
    'Delhi Daredevils'            : 'Delhi Daredevils',
    'Kings XI Punjab'             : 'Punjab Kings',
    'Pune Warriors'               : 'Rising Pune Supergiants',
    'Rising Pune Supergiant'      : 'Rising Pune Supergiants'
})
deliveries.replace({
    'Royal Challengers Bangalore' : 'Royal Challengers Bengaluru',
    'Delhi Daredevils'            : 'Delhi Daredevils',
    'Kings XI Punjab'             : 'Punjab Kings',
    'Pune Warriors'               : 'Rising Pune Supergiants',
    'Rising Pune Supergiant'      : 'Rising Pune Supergiants'
})
```

```python
# CREATE TABLES IN SQLITE DATABASE
```

```python
matches.to_sql('matches', conn, index=False, if_exists='replace')
deliveries.to_sql('deliveries', conn, index=False, if_exists='replace')

print("Tables created: 'matches' and 'deliveries'")
```

```python
# VERIFY TABLES IN DATABASE
```

```python
tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table'", conn)
print("Tables in database:")
print(tables)
```

```python
# FIRST SELECT QUERIES
```

```python
# Query 1: First 5 rows of matches
q1 = pd.read_sql_query("SELECT * FROM matches LIMIT 5", conn)
print("Matches (first 5 rows):")
display(q1)

# Query 2: First 10 rows of deliveries
q2 = pd.read_sql_query("SELECT * FROM deliveries LIMIT 10", conn)
print("\nDeliveries (first 10 rows):")
display(q2)

# Query 3: Select specific columns
q3 = pd.read_sql_query("SELECT season, team1, team2, winner FROM matches LIMIT 5", conn)
print("\nSelected columns from matches:")
display(q3)
```

```python
# CLOSE DATABASE CONNECTION
```

```python
conn.close()
print("Connection closed.")
```
